```python
import pandas as pd
import os

# def func(config):

config="/home2/ruthjohnson/atlas_qc/config"

# process config file
config_df=pd.read_csv(config, sep=':', header=None)
config_df = config_df.set_index(0)
config_df = config_df.T
new_prefix = config_df['new_prefix'].values[0].replace(" ", "")
outdir = config_df['outdir'].values[0].replace(" ", "")
 
# output from king 
king_dups_file = os.path.join(outdir, os.path.join("4_ibd", new_prefix + '.con'))
imiss_file = os.path.join(outdir, os.path.join("4_ibd", new_prefix + '.qc4.imiss'))
plink_remove_file = os.path.join(outdir, os.path.join("4_ibd", new_prefix + '.qc4.dups_remove'))

dups_df = pd.read_csv(king_dups_file, sep='\t')
dups_df = dups_df[['FID1', 'ID1', 'FID2', 'ID2']]

missing_df = pd.read_csv(imiss_file, sep=" ")
missing_df = missing_df[['FID', 'IID', 'F_MISS']]

# loop through each pair of individuals

remove_iids = [] 

for index, row in dups_df.iterrows():
    iid1 = row['ID1']
    iid2 = row['ID2']
    
    # get missing rate for both
    mr_1 = missing_df.loc[missing_df['IID'] == iid1, 'F_MISS']
    mr_2 = missing_df.loc[missing_df['IID'] == iid2, 'F_MISS']
    
    if len(mr_1.index) != 1 or len(mr_2.index) != 1:
        print("ERROR: iid not in data or multiple matching iids!")
    
    # record patient with higher missing rate
    mr_1 = mr_1.values[0]
    mr_2 = mr_2.values[0]
    
    print("Asessing missingness between: %s, %s" % (iid1, iid2))
    print("Missing rates: %.4g, %.4g" % (mr_1, mr_2))
    
    if mr_2 > mr_1:
        # remove 2nd patient
        remove_iids.append(iid2)
        print("Removing %s" % iid2)
    else: # first patient has higher missing rate
        # remove first patient
        remove_iids.append(iid1)
        print("Removing %s" % iid1)

plink_df = missing_df.loc[missing_df['IID'].isin(remove_iids)]

plink_df = plink_df[['FID', 'IID']]

plink_df.to_csv(plink_remove_file, header=None, index=False, sep=' ')
```

# PCA 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import os

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

config="/home2/ruthjohnson/atlas_qc/config"

# process config file
config_df=pd.read_csv(config, sep=':', header=None)
config_df = config_df.set_index(0)
config_df = config_df.T
new_prefix = config_df['new_prefix'].values[0].replace(" ", "")
outdir = config_df['outdir'].values[0].replace(" ", "")
batch_file = config_df['batch_file'].values[0].replace(" ", "")

outlier_file = os.path.join(outdir, os.path.join('5_pca', new_prefix + '.pca_remove'))
outlier_df = pd.read_csv(outlier_file, sep=' ')
outlier_df.columns = ['FID', 'IID']

pcs_file = os.path.join(outdir, os.path.join('5_pca', new_prefix + '.pcs'))
pcs_df = pd.read_csv(pcs_file, sep='\t')


outlier_pcs_df = pcs_df.merge(outlier_df, how='inner', on='IID')
```

```python
# vanilla PCA 

for i in range(1,10):
    plt.scatter(pcs_df['PC' + str(i)], pcs_df['PC' + str(i+1)])
    plt.xlabel("PC" + str(i))
    plt.ylabel("PC" + str(i+1))
    pc_plot_file = os.path.join(outdir, os.path.join('5_pca', new_prefix+'.pc%d_pc%d.pdf' % (i, i+1)))
    #plt.savefig(pc_plot_file)
    plt.show() 
    plt.close()
```

```python
# outlier PCA 

for i in range(1,10):
    plt.scatter(pcs_df['PC' + str(i)], pcs_df['PC' + str(i+1)])
    plt.scatter(outlier_pcs_df['PC' + str(i)], outlier_pcs_df['PC' + str(i+1)])
    plt.xlabel("PC" + str(i))
    plt.ylabel("PC" + str(i+1))
    pc_plot_file = os.path.join(outdir, os.path.join('5_pca', new_prefix+'.pc%d_pc%d.pdf' % (i, i+1)))
    #plt.savefig(pc_plot_file)
    plt.show() 
    plt.close()
```

```python
# batch info PCA 

# add batch info to PCs
batch_df=pd.read_csv(batch_file, sep=',')
batch_df['IID'] = batch_df['Harmonized BSID']
batch_df = batch_df[['IID', 'UNGC_Batch']]
batch_pcs_df = pcs_df.merge(batch_df, how='inner', on='IID')

# how many batches 
total_batches = len(batch_df['UNGC_Batch'].unique())
print("Total batches: %d" % total_batches)

for i in range(1,10):
    tmp_pcs_df = batch_pcs_df[['PC' + str(i), 'PC' + str(i+1), 'UNGC_Batch']]
    sns.scatterplot(data=tmp_pcs_df, x= 'PC' + str(i), y='PC' + str(i+1), hue='UNGC_Batch', legend=False)
    plt.xlabel("PC" + str(i))
    plt.ylabel("PC" + str(i+1))
    pc_plot_file = os.path.join(outdir, os.path.join('5_pca', new_prefix+'.pc%d_pc%d.pdf' % (i, i+1)))
    #plt.savefig(pc_plot_file)
    plt.show() 
    plt.close()
```

# add sex to fam file

```python
import pandas as pd
import matplotlib.pyplot as plt 
import os

config="/home2/ruthjohnson/atlas_qc/config"

# process config file
config_df=pd.read_csv(config, sep=':', header=None)
config_df = config_df.set_index(0)
config_df = config_df.T
new_prefix = config_df['new_prefix'].values[0].replace(" ", "")
outdir = config_df['outdir'].values[0].replace(" ", "")
ddr_sex_file = config_df['sex_file'].values[0].replace(" ", "")

fam_sex_file = os.path.join(outdir, os.path.join('6_sexcheck', new_prefix + '.qc6.fam' ))

sex_df = pd.read_csv(ddr_sex_file, sep=',')
sex_df.columns = ['SEX', 'IID']

# read in fam file from ibd step 

fam_file = os.path.join(outdir, os.path.join('4_ibd', new_prefix + '.qc4.fam'))
fam_df = pd.read_csv(fam_file, sep=' ')
fam_df.columns = ['FID', 'IID', 'MID', 'PID', 'SEX', 'PHENO']
fam_df.drop(['SEX'], axis=1, inplace=True)

fam_sex_df = fam_df.merge(sex_df, how='left', on='IID')

fam_sex_df['SEX'] = fam_sex_df['SEX'].fillna(0)
fam_sex_df['SEX'] = fam_sex_df['SEX'].astype(int)

fam_sex_df = fam_sex_df[['FID','IID', 'MID', 'PID', 'SEX', 'PHENO']]

fam_sex_df.to_csv(fam_sex_file, header=False, index=False, sep=' ')
```

# HWE histogram

```python
import pandas as pd 
import os 
import matplotlib.pyplot as plt 

%matplotlib inline

config="/home2/ruthjohnson/atlas_qc/config"

# process config file
config_df=pd.read_csv(config, sep=':', header=None)
config_df = config_df.set_index(0)
config_df = config_df.T
new_prefix = config_df['new_prefix'].values[0].replace(" ", "")
outdir = config_df['outdir'].values[0].replace(" ", "")

qc_dir=os.path.join(outdir, '3_qc')
hwe_file = os.path.join(qc_dir, new_prefix + '.qc3e.hwe')

hwe_df = pd.read_csv(hwe_file, sep=' ')

plt.hist(hwe_df['P'].values, bins=20)
plt.xlim(0,1)
plt.xlabel('p-value')
plt.title('HWE test')
outfile=os.path.join(qc_dir, new_prefix + 'qc3e.hwe.pdf')
plt.savefig(outfile)
```
